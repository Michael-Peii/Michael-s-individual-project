# Data Description #
There is two complementary tables. players.csv (196Ã—7) contains data of different players and one row represents a unique player in UBC Minecraft research sever.  
The table contains 7 columns:  
Experience: **Charactor(category)** 5 category of player's level: "Beginner, Amateur, Regular, Veteran, and Pro"  
Subscribe: **Charactor(logical)** Whether they subscribe the game related newsletter (True/False)  
HashedEmail: **Charactor** Anonymous player ID (used to connect with the hashed email within the session)  
Played_hours: **Numeric(Double)** The playing hours of different players on server (between 0 to 223.1)  
Name: **Charactor** The player's user name  
Gender: **Charactor(category)** Gender identity (7 categories: Male, female, Non-binary, Prefer not to say, Agender, Two-Spirited)  
Age: **Numeric(Double)** The age of player(from 9 to 58)  
Start time: **Charactor** human readable start time  
End time: **Charactor** human readable end time  
Original start time: **Numeric(Double)** numeric form of start time  
Original end time: **Numeric(Double)** numeric form of end time
